```python
from datetime import datetime
```

```python
class Poupanca:
    def __init__(self):
        self.saldo = 0.0
        self.taxa_rendimento_mensal = 0.005



    def calcular_rendimento(self):
        if self.saldo > 0:
            rendimento = self.saldo * self.taxa_rendimento_mensal
            self.saldo += rendimento
            return rendimento
        return 0.0

# ----------------------------------- #


class ChequeEspecial:
  def __init__(self):
    self.limite = 0.0
    self.juros_mensal = 0.05
    self.valor_ultilizado = 0.0



  def def_limite(self, novo_limite):
    if novo_limite >= 0:
      self.limite = novo_limite
      print(f'Limite de credito definido para R${novo_limite:.2f}.')
    else:
      print('ERRO: O valor não pode ser negativo.')



  def taxa_juros(self, nova_taxa_percent):
    if nova_taxa_percent >= 0:
      self.juros_mensal = nova_taxa_percent / 100.0 #converter para decimal
      print(f'Taxa de juros definida para {nova_taxa_percent}%')
    else:
      print('ERRO: O valor não pode ser negativo.')



  def juros_debitar(self):
    if self.valor_ultilizado > 0:
      juros = self.valor_ultilizado * self.juros_mensal
      return juros
    return 0.0


# ----------------------------------- #

class ContaCorrente:
  def __init__(self, nome_cliente, numero_conta):
    self.nome_cliente = nome_cliente
    self.numero_conta = numero_conta
    self.saldo = 0
    self.extrato = []

    self.poupanca = Poupanca()
    self.credito = ChequeEspecial()
    print(f"Conta criada para {self.nome_cliente}, N.º {self.numero_conta}")



  def dados(self):
    print('--- DADOS DA CONTA ---')
    print(f'Nome: {self.nome_cliente} - Nº {self.numero_conta}')
    print("--------------------------")
    print(f'Saldo: {self.saldo}')
    print(f'Poupança: {self.poupanca.saldo}')
    print("--- FIM DOS DADOS ---\n")



  def ver_extrato(self):
    agora = datetime.now()
    timestamp = agora.strftime("%d/%m/%Y %H:%M:%S")

    print("\n--- EXTRATO DA CONTA ---")
    print(f"Cliente: {self.nome_cliente}")
    print(f"Conta: {self.numero_conta}")
    print("--------------------------")

    if not self.extrato:
      print("Nenhuma transação encontrada.")
    else:
      for transacao in self.extrato:
        print(transacao)
        print("--------------------------")
      print(f"Saldo Conta Atual: R$ {self.saldo:.2f}")
      print(f"Saldo Poupança Atual: R$ {self.poupanca.saldo:.2f}")
      print(f"Limite Credito: R$ {self.saldo:.2f}")
      print(f"Fatura: R$ {self.saldo:.2f}")
      print("--- FIM DO EXTRATO ---\n")



  def depositar(self, valor):
    if valor > 0:
      self.saldo += valor

      timestamp = datetime.now().strftime("%d/%m/%Y %H:%M:%S")
      self.extrato.append(f"[{timestamp}] Depósito de R$ {valor:.2f}")

      print(f"Depósito de R$ {valor:.2f} realizado.")
    else:
      print('Valor de depósitos inválido')



  def sacar(self, valor):
    if valor < self.saldo:
      self.saldo -= valor

      timestamp = datetime.now().strftime("%d/%m/%Y %H:%M:%S")
      self.extrato.append(f"[{timestamp}] Saque de R$ {valor:.2f}")

      print(f"Saque de R$ {valor:.2f} realizado.")
    else:
      print('Valor de saque inválido')



  def transferir_poupanca(self,valor):
    if valor <= 0:
      print('Valor de tranferencia deve ser positivo')
      return

    if self.saldo >= valor:
      self.saldo -= valor
      self.poupanca.saldo += valor

      timestamp = datetime.now().strftime("%d/%m/%Y %H:%M:%S")
      self.extrato.append(f"[{timestamp}] Transferência para Poupança de R$ {valor:.2f}")

      print(f"Transferência para Poupança de R$ {valor:.2f} realizado.")
    else:
      print('Saldo insuficiente')



  def resgatar_poupanca(self,valor):
    if valor <= 0:
      print('Valor de tranferencia deve ser positivo')
      return

    if self.poupanca.saldo >= valor:
      self.poupanca.saldo -= valor
      self.saldo += valor

      timestamp = datetime.now().strftime("%d/%m/%Y %H:%M:%S")
      self.extrato.append(f"[{timestamp}] Resgate da Poupança de R$ {valor:.2f}")

      print(f"Resgate da Poupança de R$ {valor:.2f} realizado.")
    else:
      print('Saldo insuficiente')



  def aplicar_rendimento_poupanca(self):
    valor_rendimento = self.poupanca.calcular_rendimento()

    if valor_rendimento > 0:
      timestamp = datetime.now().strftime("%d/%m/%Y %H:%M:%S")
      self.extrato.append(f"[{timestamp}] Rendimento da poupança de R$ {valor_rendimento:.2f}")

      print(f"Rendimento de R$ {valor_rendimento:.2f} realizado.")
    else:
      print('Saldo insuficiente')



  def def_limite(self, valor_limite):
    self.credito.def_limite(valor_limite)



  def taxa_juros(self, taxa_percent):
    self.credito.taxa_juros(taxa_percent)



  def juros_debitar(self):
    valor_juros = self.credito.juros_debitar()

    if valor_juros > 0:
      self.saldo -= valor_juros

      timestamp = datetime.now().strftime("%d/%m/%Y %H:%M:%S")
      self.extrato.append(f"[{timestamp}] Débito Juros Ch. Especial R${valor_juros:.2f}")
      print(f"Debito de R$ {valor_juros:.2f} realizado.")
    else:
      print('Nenhum juro a ser debitado.')
```
